# Image compression analysis

## Imports

```python
import os
import re
import matplotlib.pyplot as plt
import pandas as pd
```

## Celebrities

```python
DIR = "celebrities/"
pattern = re.compile(
    r"Compression ratio:\s*([\d.]+)\s*\|\s*Compression MSE:\s*([\d.]+)\s*\|\s*Compression PSNR:\s*([\d.]+)"
)
data = []

for filename in os.listdir(DIR):
    filepath = DIR + filename
    if not os.path.isfile(filepath):
        continue

    f = open(filepath, "r")
    lines = f.read()
    f.close()

    matches = pattern.findall(lines)
    if not matches:
        raise ValueError("Invalid pattern")

    for m in matches:
        ratio, mse, psnr = map(float, m)
        data.append({"ratio":ratio, "mse":mse, "psnr":psnr})

df = pd.DataFrame(data)
df
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

metrics = {
    "Ratio": df["ratio"],
    "MSE": df["mse"],
    "PSNR": df["psnr"],
}

for name, values in metrics.items():
    fig, ax = plt.subplots(figsize=(6, 4), dpi=150)

    ax.boxplot(
        values,
        vert=True,
        patch_artist=True,
        boxprops=dict(facecolor="#4B8BBE", color="#306998", alpha=0.5),
        medianprops=dict(color="#FFD43B", linewidth=2),
        whiskerprops=dict(color="#306998"),
        capprops=dict(color="#306998"),
        flierprops=dict(marker="o", color="#FF6F61", markersize=5, alpha=0.7),
    )

    ax.set_title(f"Celebrities {name} Distribution", fontsize=15, weight="bold", pad=12)
    ax.set_ylabel(name, fontsize=11)
    ax.set_xticks([])
    ax.grid(alpha=0.3)

    # Quitar bordes superiores y derechos
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.tight_layout()
    plt.show()
```
